```python
df['age'] = pd.to_numeric(df['age'] , errors='coerce')
df['bp'] = pd.to_numeric(df['bp'] , errors='coerce')
df['bgr'] = pd.to_numeric(df['bgr'] , errors='coerce')
df['bu'] = pd.to_numeric(df['bu'] , errors='coerce')
df['sc'] = pd.to_numeric(df['sc'] , errors='coerce')
df['sod'] = pd.to_numeric(df['sod'] , errors='coerce')
df['pot'] = pd.to_numeric(df['pot'] , errors='coerce')
df['hemo'] = pd.to_numeric(df['hemo'] , errors='coerce')
df['pcv'] = pd.to_numeric(df['pcv'] , errors='coerce')
df['wbcc'] = pd.to_numeric(df['wbcc'] , errors='coerce')
df['rbcc'] = pd.to_numeric(df['rbcc'] , errors='coerce')


```

```python
def fill_values(dataframe):
    df['age'] =  dataframe["age"].fillna(int(df['age'].mean()))
    df['bp'] =  dataframe["bp"].fillna(int(df['bp'].mean()))
    df['bgr'] =  dataframe["bgr"].fillna(int(df['bgr'].mean()))
    df['bu'] =  dataframe["bu"].fillna(int(df['bu'].mean()))
    df['sc'] =  dataframe["sc"].fillna(int(df['sc'].mean()))
    df['sod'] =  dataframe["sod"].fillna(int(df['sod'].mean()))
    df['pot'] =  dataframe["pot"].fillna(int(df['pot'].mean()))
    df['hemo'] =  dataframe["hemo"].fillna(int(df['hemo'].mean()))
    df['pcv'] =  dataframe["pcv"].fillna(int(df['pcv'].mean()))
    df['wbcc'] =  dataframe["wbcc"].fillna(int(df['wbcc'].mean()))
    df['rbcc'] =  dataframe["rbcc"].fillna(int(df['rbcc'].mean()))

    df['sg'] = dataframe["sg"].fillna(str(df['sg'].mode()[0]))
    df['al'] = dataframe["al"].fillna(str(df['al'].mode()[0]))
    df['su'] = dataframe["su"].fillna(str(df['su'].mode()[0]))
    df['rbc'] = dataframe["rbc"].fillna(str(df['rbc'].mode()[0]))
    df['pc'] = dataframe["pc"].fillna(str(df['pc'].mode()[0]))
    df['pcc'] = dataframe["pcc"].fillna(str(df['pcc'].mode()[0]))
    df['ba'] = dataframe["ba"].fillna(str(df['ba'].mode()[0]))

    df['htn'] = dataframe["htn"].fillna(str(df['htn'].mode()[0]))
    df['dm'] = dataframe["dm"].fillna(str(df['dm'].mode()[0]))
    df['cad'] = dataframe["cad"].fillna(str(df['cad'].mode()[0]))
    df['appet'] = dataframe["appet"].fillna(str(df['appet'].mode()[0]))
    df['pe'] = dataframe["pe"].fillna(str(df['pe'].mode()[0]))
    df['ane'] = dataframe["ane"].fillna(str(df['ane'].mode()[0]))
    df['class'] = dataframe["class"].fillna(str(df['class'].mode()[0]))

    
```
